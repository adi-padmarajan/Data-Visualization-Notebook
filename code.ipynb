# Plot Methods
1. plt.plot(x, y) - Line Plot
2. plt.bar(x, height) - Vertical Bar Chart
3. plt.barh(y, width) - Horizontal Bar Chart
4. plt.scatter(x, y) - Scatter Plot
5. plt.py(data) - Pie Chart
6. plt.hist(data, bins) - Histogram
7. plt.boxplot(data) - Box Plot

# Bar Charts
A bar plot (or bar chart) is a graph that displays categorical data using rectangular bars. Each bar's height (or length, in horizontal bars) represents a value.

```python
movies = ["The Dark Kinght", "The Hurt Locker", "The King's Speech", "The Artist", "Argo"]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170,427, 133, 232]

plt.bar(years, oscar_revenue, color = "purple")
plt.title("Revenue in each year") 
plt.xlabel("Year")
plt.ylabel("Revenue ($M)")

# plt.text()
# (x-axis pos, y-axis pos, text, alignment)

'''
We can increase the size of the graph using xlim and ylim
'''

for i in range(len(years)):
    plt.text(years[i], oscar_revenue[i] + 10, str(oscar_revenue[i]), ha = "center")

plt.ylim(0, max(oscar_revenue) + 100)
```

# Compare 2 Sets of Data 
We customize width. Width controls how thick each bar is. Usually ranges from 0 to 1. [Default value is 0.8]

```python
import numpy as np

oscar_movies = ["The Dark Kinght", "The Hurt Locker", "The King's Speech", "The Artist", "Argo"]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170,427, 133, 232]
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]

# We need to manipulate the width of the bars to display side by  (Default width is 0.8)
x = np.arange(len(years)) 
# print(x) [0, 1, 2, 3, 4]

width = 0.4 
plt.bar(x - width/2, oscar_revenue, width, label = "Oscar Movies") # Shifting to left
plt.bar(x + width/2, non_oscar_revenue, width, label = "Non-Oscar Movies") # Shifting to right 

plt.title("Oscar vs Non-Oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue ($M)")
plt.legend()

plt.xticks(x, years) # to set the years as the labels
plt.show()
```
