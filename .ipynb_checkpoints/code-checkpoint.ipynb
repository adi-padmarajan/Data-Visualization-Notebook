```python
import matplotlib.pyplot as plt
```

```python
X = [1, 2, 3, 4]
Y = [5, 6, 7, 8]

# Line Chart
plt.plot(X, Y) 
```

```python
oscar_movies = [
    "The Dark Kinght",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170,427, 133, 232]

# Plot
plt.plot(years, oscar_revenue)

'''
title()
xlabel()
ylabel()
legend()
'''

plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Non-Oscar Winnder Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]

plt.plot(years, oscar_revenue, label = "Oscar Movies")
plt.plot(years, non_oscar_revenue, label = "Non-Oscar Movies")

plt.title("Oscar Movies Revenue Vs. Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()
```
