```python
import matplotlib.pyplot as plt
```

```python
oscar_movies = [
    "The Dark Kinght",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170,427, 133, 232]

# Plot
plt.plot(years, oscar_revenue)

'''
title()
xlabel()
ylabel()
legend()
'''

plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Non-Oscar Winnder Movies
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]

plt.plot(years, oscar_revenue, label = "Oscar Movies")
plt.plot(years, non_oscar_revenue, label = "Non-Oscar Movies")

plt.title("Oscar Movies Revenue Vs. Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()
```

```python
'''
Format Strings 
A format string consists of a part for color, marker and line.

fmt = '[marker][line][color]'
- color
- marker
- linestyle
- linewidth
'''

oscar_movies = ["The Dark Kinght", "The Hurt Locker", "The King's Speech", "The Artist", "Argo"]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170,427, 133, 232]
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]

plt.plot(years, oscar_revenue,"o--r",label = "Oscar Movies")
plt.plot(years, non_oscar_revenue, "o-g",label = "Non-Oscar Movies")

plt.title("Oscar Movies Revenue Vs. Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()

'''
- style.available
- style.use()
- grid()
- tight_layout()
- savefig("abc.png")
'''

print(plt.style.available)
plt.show()
```
